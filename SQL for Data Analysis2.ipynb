```python
import pandas as pd 
import pandasql as ps
```

```python
df=pd.read_csv('titanic.csv')
df.head()
```

```python
# Aggergation 
ps.sqldf("""

select
    df.Embarked,
    avg(df.age) as avg_age,
    sum(df.age) as tot_age,
    min(df.age) as min_age,
    max(df.age) as max_age,
    count(df.PassengerId) as tot_passengers,
    count(distinct df.Ticket) as dist_tickets
FROM df
GROUP by df.Embarked

""")
```

```python
ps.sqldf("""

select
    df.Embarked,
    df.Sex,
    avg(df.age) as avg_age,
    sum(df.age) as tot_age,
    min(df.age) as min_age,
    max(df.age) as max_age,
    count(df.PassengerId) as tot_passengers,
    count(distinct df.Ticket) as dist_tickets
FROM df
GROUP by df.Embarked,
    df.Sex

""")
```

```python
ps.sqldf("""

select
    df.Embarked,
    df.Sex,
    avg(df.age) as avg_age,
    sum(df.age) as tot_age,
    min(df.age) as min_age,
    max(df.age) as max_age,
    count(df.PassengerId) as tot_passengers,
    count(distinct df.Ticket) as dist_tickets
FROM df

where age >= 18 and age <= 60

GROUP by df.Embarked,
    df.Sex

""")
```

```python
ps.sqldf("""

select
    df.Embarked,
    df.Sex,
    avg(df.age) as avg_age,
    sum(df.age) as tot_age,
    min(df.age) as min_age,
    max(df.age) as max_age,
    count(df.PassengerId) as tot_passengers,
    count(distinct df.Ticket) as dist_tickets
FROM df

where age >= 18 and age <= 60

GROUP by df.Embarked,
    df.Sex
    
order by df.Embarked desc,
    df.Sex asc

""")
```

```python
ps.sqldf("""

select
    df.Embarked,
    df.Sex,
    avg(df.age) as avg_age,
    sum(df.age) as tot_age,
    min(df.age) as min_age,
    max(df.age) as max_age,
    count(df.PassengerId) as tot_passengers,
    count(distinct df.Ticket) as dist_tickets
FROM df

where age >= 18 and age <= 60

GROUP by df.Embarked,
    df.Sex
    
having avg_age > 35
    
order by df.Embarked desc,
    df.Sex desc

""")
```

# JOIN

```python
cities = pd.DataFrame({'code' : ['S','C','L'],'city':['Southampton','Cherbourg','London']})
cities
```

```python
ps.sqldf("""

select
    c.city,
    df.Embarked,
    df.Sex,
    avg(df.age) as avg_age,
    sum(df.age) as tot_age,
    min(df.age) as min_age,
    max(df.age) as max_age,
    count(df.PassengerId) as tot_passengers,
    count(distinct df.Ticket) as dist_tickets
FROM df

inner join cities as c
    on df.Embarked = c.code
    
group by df.Embarked,
    df.Sex 

""")
```

```python
ps.sqldf("""

select
    c.city,
    df.Embarked,
    df.Sex,
    avg(df.age) as avg_age,
    sum(df.age) as tot_age,
    min(df.age) as min_age,
    max(df.age) as max_age,
    count(df.PassengerId) as tot_passengers,
    count(distinct df.Ticket) as dist_tickets

FROM df
left join cities as c
    on df.Embarked = c.code
    
group by df.Embarked,
    df.Sex 

""")
```

```python
ps.sqldf("""

select
    c.city,
    df.Embarked,
    df.Sex,
    avg(df.age) as avg_age,
    sum(df.age) as tot_age,
    min(df.age) as min_age,
    max(df.age) as max_age,
    count(df.PassengerId) as tot_passengers,
    count(distinct df.Ticket) as dist_tickets

FROM cities as c
left join df
    on df.Embarked = c.code
    
group by df.Embarked,
    df.Sex 

""")
```
